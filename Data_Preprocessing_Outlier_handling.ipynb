```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('/content/beer-servings.csv')
df.head()
```

```python
df.columns
```

```python
# IQR Method
Q1 = df['beer_servings'].quantile(0.25)
Q3 = df['beer_servings'].quantile(0.75)
IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)
```

```python
# outlier threshold
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
print(upper_bound)
print(lower_bound)
```

```python
# identify outliers using IQR

outliers_IQR = df[~((df['beer_servings'] >= lower_bound) & (df['beer_servings'] <= upper_bound))]
outliers_IQR
```
